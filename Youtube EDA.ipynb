## Data Cleaning

```python
## Checking for duplicates in the data

df.duplicated().sum()
```

```python
## Checking for null values in the data

df.isnull().sum()
```

```python
## Removing columns that will not be used for Data Analysis

df.drop(['Gross tertiary education enrollment (%)', 'Population', 'Unemployment rate', 'Urban_population', 'Latitude', 'Longitude', 'subscribers_for_last_30_days'], axis = 1, inplace = True)
```

```python
## Filling null values of columns such as "category","Abbreviation" and "channel_type" with "Not Available"

df[['category', 'Abbreviation', 'channel_type']] = df[['category', 'Abbreviation', 'channel_type']].fillna('Not Available')
```

```python
## Drop any missing values

df.dropna(inplace = True)
```

```python
## Checking the data types 

df.dtypes
```

```python
## Changing the data type of the created_date column to date data type

df.created_date = pd.to_datetime(df.created_date)
```

### Category wise Average Subscribers

* Shows: This category has the highest average number of subscribers, exceeding 40 million, indicating its strong popularity and wide appeal to a large audience.
* Trailers: Close behind, the Trailers category also boasts a significant average subscriber count, likely reflecting the broad interest in movie, TV, or game trailers.
* Film & Animation: This category maintains a solid average of over 30 million subscribers, highlighting the appeal of creative and visual content like animated films or short films.
* Nonprofits & Activism: Despite being focused on social causes, this category also commands a respectable average of around 30 million subscribers, indicating strong engagement in content that promotes social change or activism.
* Education: The Education category has a similar average subscriber count to Nonprofits & Activism, suggesting that informative, instructional content has a significant following, though slightly lower compared to Shows and Trailers.
* Insights: Shows and Trailers appear to be the most popular categories in terms of average subscribers, likely due to their mass appeal and entertainment value.
* Categories like Nonprofits & Activism and Education, while slightly lower in average subscribers, still demonstrate strong engagement, particularly in niches that focus on social impact or educational content.
* This data suggests that content creators focusing on Shows or Trailers have the potential to reach a larger audience, while those in Nonprofits & Activism and Education can still achieve significant viewership with high-value content in those niches

### Correlation between subscribers, video views and earnings

```python
colors = ["#9A3B3B","#C08261","#E2C799","#FFF1E5"]
custom_cmap = LinearSegmentedColormap.from_list("mycmap",colors)
plt.subplots(figsize=(6,4))
corr = new_df[['subscribers', 'video views','lowest_monthly_earnings','highest_monthly_earnings']].corr()
sns.heatmap(corr, annot = True, cmap = custom_cmap, linewidths = 0.6, linecolor = "black")
plt.title('Correlation between variables')
plt.show()
```

* This visual is a correlation matrix, which shows the relationships between different variables using correlation coefficients.
* The values range from -1 to 1. 1 indicates a perfect positive correlation (variables move in the same direction). -1 indicates a perfect negative correlation (variables move in opposite directions). 0 means no correlation.
* Subscribers & Video Views: Correlation of 0.79. This is a strong positive correlation, indicating that as the number of subscribers increases, the number of video views also tends to increase. Video Views & Monthly Earnings (both lowest and highest): Correlation of 0.64. This suggests a moderate positive relationship, meaning that higher video views are associated with higher monthly earnings. Subscribers & Monthly Earnings (both lowest and highest): Correlation of 0.5. This indicates a moderate positive relationship, suggesting that an increase in subscribers is moderately associated with an increase in monthly earnings. Lowest & Highest Monthly Earnings: Correlation of 1
* The strong correlation between subscribers and video views suggests that increasing subscriber count can significantly boost video views, which is expected as subscribers are usually more engaged viewers.
* The moderate correlation between video views and monthly earnings implies that more video views are generally associated with higher earnings, which makes sense in most ad-revenue-based platforms.
* This matrix helps identify which variables are most related to each other and could inform decisions on where to focus efforts for growth or understanding dependencies between different metrics.
